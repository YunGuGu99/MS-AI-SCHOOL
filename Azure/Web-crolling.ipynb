```python
import bs4
import requests
```

```python
url = 'https://search.shopping.naver.com/search/category/100005307'
result = requests.get(url).text
result
```

```python
bsObj = bs4.BeautifulSoup(result, 'html.parser')
```

```python
items = bsObj.find_all('a', {'class': 'basicList_link__JLQJf'})   #find - 찾기
```

```python
for item in items:
    print(item.text)
```

```python
print(items[0].text)
print(items[1].text)
print(items[2].text)
print(items[3].text)
```
